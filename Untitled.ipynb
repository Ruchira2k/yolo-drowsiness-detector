```python
import uuid ##unique identifier
import os
import time
import random
```

```python
IMAGES_PATH = os.path.join('data','images')
labels = ['awake', 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img))
        
        ret, frame = cap.read()
        
        image_name = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(image_name, frame)
        cv2.imshow('Image collection', frame)
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
        
```

# Test

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp6/weights/last.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0)
ringer = False
alarm_tone = vlc.MediaPlayer(f"file:///Radar.mp3")
while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
#     print(results.xywh[0])
    if len(results.xywh[0]) > 0:
        dconf = results.xywh[0][0][4]
        dclass = results.xywh[0][0][5]
#         alarm_tone = vlc.MediaPlayer(f"file:///Radar.mp3")
        if dconf > 0.6 and dclass == 16:
            ringer = True
            print(ringer)
            alarm_tone.play()
#         
        if dconf > 0.7 and dclass == 15:
            ringer = False
            alarm_tone.stop()
            print(ringer)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        alarm_tone.stop()
        break
        
cap.release()
cv2.destroyAllWindows()
```
